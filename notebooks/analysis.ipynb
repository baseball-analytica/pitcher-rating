```python
import pybaseball
import pandas as pd
import numpy as np
```

```python
pybaseball.cache.enable()
```

```python
pitcher_data = pybaseball.pitching_stats(2025)

print(pitcher_data.head())
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(pitcher_data["K%"])
plt.title("Distribution of K%, 2025 qualified pitchers")

plt.show()
```

```python
plt.hist(pitcher_data["BB%"])
plt.title("Distribution of BB%, 2025 qualified pitchers")

plt.show()
```

```python
plt.hist(pitcher_data["HBP"] / pitcher_data["TBF"])
plt.title("Distribution of HBP%, 2025 qualified pitchers")

plt.show()
```

```python
plt.hist(pitcher_data["HR"] / pitcher_data["TBF"])
plt.title("Distribution of HR%, 2025 qualified pitchers")

plt.show()
```
